```python
#130
import requests
btc=requests.get("https://api.bithumb.com/public/ticker/").json()['data']

btc["opening_price"]
btc["closing_price"]
btc["min_price"]
btc["max_price"]

byun=int(btc["max_price"])-int(btc["min_price"])
siga=int(btc["opening_price"])
max=int(btc["min_price"])
if (byun+siga) > max:
  print("상승창")
else:
  print("하락창")
```
